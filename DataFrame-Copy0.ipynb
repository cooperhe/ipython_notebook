# 1.Import and merge temp

```python
for i in range(16):
    
    Dt_i = pd.read_excel('051101.xlsx', i, header=10)
    Dt_i.columns=[1,2,3]
    Dt_i = pd.DataFrame(Dt_i, columns=[1,2])
    Dt_i.columns=['time',i]
    
    if i ==0:
        Data = Dt_i
    else:
        Data = pd.merge(Data,Dt_i, how='outer')
   
Data
```

# 2.find the zero point

```python
Dt=Data

for x in range(len(Dt.index)):
    if Dt[0][x+1]-Dt[0][x]>0:
        if Dt[0][x+2]-Dt[0][x+1]>0:
            if Dt[0][x+3]-Dt[0][x+2]>0:
                if Dt[0][x+4]-Dt[0][x+3]>0:
                    if Dt[0][x+5]-Dt[0][x+4]>0:
                        if Dt[0][x+5]-Dt[0][x]>3:
                            break

for i in range(len(Dt.index)):
    if i < len(Dt.index)-x:
        for j in range(len(Dt.columns)-1):
            Dt[j][i]=Dt[j][i+x]
    else:
        Dt = Dt.drop(Dt.index[i:])
    
Dt
```
