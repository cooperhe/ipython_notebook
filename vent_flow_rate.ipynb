# Why Horizontial Vent Flow? -- Motivation

**Vent flow calaulation is quite importment in fires, since it determines the ventilation condition of the fire in rooms. But as wrote by Howard W. Emmons in the SFPE handbook, "Unlike nonbuoyant flows through orifices or flow through vents in a vertical wall, very little quantitative work has been done on flow through vents in horizontal (floors or flat roofs) or slightly sloped (inclined roofs) surfaces". he also clarifid the present status of our knowledge only provided flow calculation formulas of unknown accuracy in lieu of nothing. Studys on horizontial vent flow under fire conditions are in great need but it is always not an easy job.**

###Existing vent flow theory
  <img src="sketch.png" alt="Drawing" style="width: 200px;"/>
  Consider the flow through a hole in a horizontal surface. While the pressure difference is significate, the velocity and flow rate can be determined from the pressure drop through a distance of holes using Bernoulli Equation. 
  <center>$$v=[\frac{2}{\rho}(\triangle \rho gh+\triangle p)]^{1/2}$$<center>
  
  However, while it comes to a fire-induced ceiling vent flow, both pressure difference and density difference through the vent are small, leading the vent flow nearly zero. the interface between the lower (hot) and upper (cold) gases becomes flat and is unstable. The unidirectional flow is replaced by simultaneous up and down flows usually oscillating in time and location. In addition, ceiling or roof vents in fires are usually irregular in shape and have a length to “diameter” ratio of 0.13 or less. There are some studies of rooms with a ceiling hole with a fire under the vent. These supply interesting fire data but are not useful as horizontal vent studies, since the results do not include adequate orifice pressure and flow measurements.
  
  The recent calculation of ceiling vent flow in fires was still based on Cooper's theory back to 1980s. Cooper considered that there will be a critical value of the pressure difference to distinguish between unidirectional flow regime and a mixed flow regime,
  <center>$$\triangle P_{cri} = 0.2427(1+\frac{1}{2}ε) exp(1.1072ε) (4g \triangle ρ D)$$<center>
  
  where $ε = -\triangle \rho / \overline \rho $ and is related to the Atwood number, $g$ is the acceleration due to gravity, $\triangle ρ$ is the density difference and $D$ is the equivalent diameter of the ceiling vent.
  
  Correspondingly, the critical flow rate is
  <center>$$\dot V_{cri} = \frac{1}{4} \pi D^2 (2gD \triangle \rho / \rho_0 )0.1754 exp (0.5536ε) $$<center>
  
  When the pressure difference across the ceiling vent $\triangle P_{0} ≥ \triangle P_{cri}$, there will be a unidirectional flow, and the flow rate can be expressed by
  <center>$$ \dot V = \big\{ 1−C_2^2 +[C_2^42 +C_1^2 (\triangle P_o /\triangle P_{cri} −1)]^\frac{1}{2} \big\} \dot{ V_{cri} } $$ <center>
  
  where $c_1 ≈ 3.37$ and $c_2 ≈ 1.045$.
  
  When $0 ≤ \triangle P_{0} ≤ \triangle P_{cri}$, there will be a bidirectional flow, the net volume flowrate across the ceiling vent can be calculated by
  <center>$$ \dot{V_{net}} = \dot{V_{cri}} \{ M −[1+(M^2 −1)(1−\triangle P_o /\triangle P_{cri}  ]^ \frac{1}{2} \}/(M-1)$$<center>
  
  where $M = 9.1$.
  
  The volume flow rate into the compartment can be expressed as
  <center>$$ \dot V_{u->l} = [(1+ \frac{m_3}{2})(1− \triangle P_o /\triangle P_{cri} )^2 -(2+\frac{m_3}{2})(1− \triangle P_o / \triangle P_{cri})]^2 \dot V_{ex.max} $$<center>
  
  where $m_3$ is 0.7070 and $\dot V_{ex,max}$ is the maximum volume
exchange rate while $\triangle P_o =0$. Epstein have measured the transfer of fluid from the lower to the upper chambers using water below and a brine above (density ratio 1.1 or less). They found that, at zero net volume flow (the lower chamber was closed except for a ceiling hole 2 in. in diameter or less), the fluid transfer to the upper chamber was
<center>$$ \dot V_{ex,max} = 0.055 [D^5 g (\triangle \rho / \overline \rho )]^{1/2}$$<center>
  
  
  At fire conditions, the $ε = -\triangle \rho / \overline \rho =−2(T_g −T_0)/(T_g +T_0)$

  We can get the critical pressure and critical flow rate as the function of Gas temperature as:
  <center>$$\triangle P_{cri} = 0.2427(1+\frac{1}{2}ε) exp(1.1072ε) (4g \triangle ρ D)$$<center>
  <center>$$\dot V_{cri} = \frac{1}{4} \pi D^2 (2gD \triangle \rho / \rho_0 )0.1754 exp (0.5536ε) $$<center>

```python
T0=25.
Tg=np.linspace(25, 1000, 500)
e=2.*(T0-Tg)/(T0+Tg)
D=0.319
Pcri=0.2427*(1.+0.5*e)*exp(1.1072*e)*(4*9.8*D*1.21*(Tg-T0)/(Tg+273))
Vcri=1.0/4.0*np.pi*D**2*(2*9.8*D*(Tg-T0)/(Tg+273))*0.1754*exp(0.5536*e)
    
figure(figsize=(12,4))

subplot(1,2,1)
plot(Tg,Pcri,'r-')
xlabel('$T_g / ^oC$')
ylabel('$ P_{cri} / Pa$')
title('Critical Pressure $P_{cri}$')
grid()

subplot(1,2,2)
plot(Tg,Vcri,'r-')
xlabel('$T_g / ^oC$')
ylabel('$ V_{cri} / m^3 / s$')
title('Critical Flow Rate')
grid()
```

From the experiment, we know the fire can burn for a relative long time. Fresh air should have flew in the compartment throught the vent. There will be a bidirectional flow, and $0 ≤ \triangle P_{0} ≤ \triangle P_{cri}$.Thus:
  <center>$$ \dot V_{net} = \dot V_{cri} \{ M −[1+(M^2 −1)(1−\triangle P_o /\triangle P_{cri}  )]^ \frac{1}{2} \}/(M-1)$$<center>

```python
from IPython.html.widgets import interact
T0=25.
Tg=np.linspace(25, 1000, 500)
e=2.*(T0-Tg)/(T0+Tg)
D=0.319
Pcri=0.2427*(1.+0.5*e)*exp(1.1072*e)*(4*9.8*D*1.21*(Tg-T0)/(Tg+273))
Vcri=1.0/4.0*np.pi*D**2*(2*9.8*D*(Tg-T0)/(Tg+273))*0.1754*exp(0.5536*e)
def plot_Vnet(P0_to_Pcri):
    M=9.1
    Vnet=Vcri*(M-(1.+(M**2-1.)*(1.-P0_to_Pcri))**0.5)/(M-1.)
    plot(Tg,Vnet,'r-', label='Vnet')
    plot(Tg,Vcri,'k-', label='Vcri')

    xlabel('$T_g / ^oC$')
    ylabel('$ V_{net} / m^3 / s$')
    title('Net Flow Rate')
    legend(loc='lower right')
    grid()
    
interact(plot_Vnet, P0_to_Pcri=(0, 1, 0.1));
```

<center>$$ \dot V_{ex,max} = 0.055 [D^5 g (\triangle \rho / \overline \rho )]^{1/2}$$<center>

```python
from IPython.html.widgets import interact
T0=25.
Tg=np.linspace(25, 1000, 500)
e=2.*(T0-Tg)/(T0+Tg)
D=0.319
Pcri=0.2427*(1.+0.5*e)*exp(1.1072*e)*(4*9.8*D*1.21*(Tg-T0)/(Tg+273))
Vcri=1.0/4.0*np.pi*D**2*(2*9.8*D*(Tg-T0)/(Tg+273))*0.1754*exp(0.5536*e)
Vmax=0.055*sqrt(D**5.*9.8*(-e))
m3=0.7070

def plot_V(P0_to_Pcri):
    M=9.1
    Vnet=Vcri*(M-(1.+(M**2-1.)*(1.-P0_to_Pcri))**0.5)/(M-1.)
    Vin=Vmax*((1.+m3/2.)*(1.-P0_to_Pcri)**2-(2.+m3/2.)*(1.-P0_to_Pcri))**2
    plot(Tg,Vnet/Vin,'b-', label='Vout')

    xlabel('$T_g / ^oC$')
    ylabel('$ V_{in \, or \, out} / m^3 / s$')
    title('in/out Flow Rate')
    legend(loc='lower right')
    grid()
    
interact(plot_V, P0_to_Pcri=(0, 1, 0.1));
```

On the other aspect, the inflow and outflow should obey the mass conservation equation:
<center>$$\dot m_{in}+\dot m_{f}=\dot m_{out}$$<center>

We know the fuel flow rate $\dot m_f$
At case of fuel flow rate of 15SLPM, $\dot m_f=4.575* 10^{-4} kg/s$. While at steady state, the smallest in flow air should still supply the combustion. The stoichiometric value is found from:
$$C_3H_8 + 5(O_2+3.76N_2)  \to 3 CO_2+ 4 H_2O + 18.8 N_2$$

The mass ratio of inflow air/fuel=686.4/44, so $\dot m_f$ is much smaller than $\dot m_{in}$, so we get $\dot m_{in} \approx \dot m_{out}$. 

Thus, we should have:
<center>$$\frac{\dot V_{net}}{\dot V_{in}}=\frac{\dot V_{out} -\dot V_{in} }{\dot V_{in}}=\frac{T_g-T_0}{T_0}\, or \, \frac{\dot V_{out} }{\dot V_{in}}=\frac{T_g}{T_0}$$ <center>

And from the simulation, we take $T_g$as the average compartment temperature. We get $T_g=280^oC \, or\,  553K$

Then, we can solve $\triangle P_o /\triangle P_{cri}$ from all the equations above for a given Tg. The results are shown below:

```python
%pylab inline

T0=25.
P=range(200)
Vi=range(200)
Vo=range(200)
i=0
for Tg in np.linspace(25,1000,200):
    e=2.*(T0-Tg)/(T0+Tg)
    D=0.319
    Pcri=0.2427*(1.+0.5*e)*exp(1.1072*e)*(4*9.8*D*1.21*(Tg-T0)/(Tg+273))
    Vcri=1.0/4.0*np.pi*D**2*(2*9.8*D*(Tg-T0)/(Tg+273))*0.1754*exp(0.5536*e)
    Vmax=0.055*sqrt(D**5.*9.8*(-e))
    m3=0.7070
    M=9.1


    for P0_to_Pcri in np.linspace(0,1.0,200):
        Vnet=Vcri*(M-(1.+(M**2-1.)*(1.-P0_to_Pcri))**0.5)/(M-1.)
        Vin=Vmax*((1.+m3/2.)*(1.-P0_to_Pcri)**2-(2.+m3/2.)*(1.-P0_to_Pcri))**2
        if 0<(Vnet+Vin)/Vin-(Tg+273)/(T0+273)<1:
            P[i]=P0_to_Pcri
            Vi[i]=Vin
            Vo[i]=Vin+Vnet           
            break
    i=i+1
    
figure(figsize=(12,4))
subplot(1,2,1)
plot(np.linspace(26,1000,200),P,'b-', label='P0_to_Pcri')
xlabel('$T_g / ^oC$')
ylabel('$ P_0 / P_{cri}$')
title('P0_to_Pcri predicted by Coopy')
legend(loc='lower center')
grid()
subplot(1,2,2)
plot(np.linspace(26,1000,200),Vi,'b-', label='Cooper-inflow rate')
plot(np.linspace(26,1000,200),Vo,'r-',label='Cooper-outflow rate')
xlabel('$T_g / ^oC$')
ylabel('$ flow rate / m^3/s$')
title('flow predicted by Coopy')
legend(loc='lower center')
grid()
```

###Theorical Analysis
**Cooper'theory has large error while the strong fire plume effects the vent flow, we need to introduce a new nomdemissional number to determine influence**

While the both sides of the vent are almost quiesce, the horizental vent flow are mainly determined by the buoyancy force peoduce by the density different of fliud on the two sides, whcih can be presented as:
$$F_{buoyancy}=(\rho_0-\rho_g)gL^3$$

While the outer gas flow reaches the vent, the inertial force can be presented as:
$$F_{inertial}=\rho v^2 L^2$$

We can redefine the Froud as the ratio of buoyancy force to inertial force:
$$Fr=\frac{buoyancy \, force}{inertial \, force}=\frac{(\rho_0-\rho_g)gL}{\rho v^2}$$
