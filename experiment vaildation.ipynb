```python
#read temperature data from file and merge it into a large DataFrame
for i in range (40):
    Temp_i = pd.read_excel('/home/qize/CVC_experiment/Result/20140511/temp/051101.xlsx', sheetname=i, header=10)
    Temp_i.columns = ['time', Temp_i.columns[1][42:],'c']
    Temp_i = pd.DataFrame(Temp_i, columns=['time', Temp_i.columns[1]])
    
    if i ==0:
        Data = Temp_i
    else:
        Data = pd.merge(Data,Temp_i, how='outer')
#get the DataFrame Data
```

```python
#find the zero point and delete the date before ignition
Dt=Data

for x in range(len(Dt.index)):
    if Dt['Mod5_ai9'][x+1]-Dt['Mod5_ai9'][x]>0:
        if Dt['Mod5_ai9'][x+2]-Dt['Mod5_ai9'][x+1]>0:
            if Dt['Mod5_ai9'][x+3]-Dt['Mod5_ai9'][x+2]>0:
                if Dt['Mod5_ai9'][x+4]-Dt['Mod5_ai9'][x+3]>0:
                    if Dt['Mod5_ai9'][x+5]-Dt['Mod5_ai9'][x+4]>0:
                        if Dt['Mod5_ai9'][x+5]-Dt['Mod5_ai9'][x]>3:
                            break

Dt = Dt.drop(Dt.index[:x])
Dt.index=range(Dt.index.size)
#find the zero point is x, and update the DataFrame as Dt
```
