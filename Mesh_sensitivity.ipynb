##     1. HRR

```python
HRR20mm = pd.read_csv('/home/qize/FDS_simulation/Copy_sever/mesh_sensitivity_study/20mm/2D_CVC_V40_15SLPM_hrr.csv', header=1)
HRR10mm = pd.read_csv('/home/qize/FDS_simulation/Copy_sever/mesh_sensitivity_study/10mm/2D_CVC_V40_15SLPM_hrr.csv', header=1)
HRR5mm = pd.read_csv('/home/qize/FDS_simulation/Copy_sever/mesh_sensitivity_study/5mm/2D_CVC_V40_15SLPM_hrr.csv', header=1)

```

```python
plot(HRR20mm.Time,HRR20mm.HRR, 'ko', markersize=1)

# average from 2 before until 2 after (window size = 5)
plot(HRR20mm.Time, pd.rolling_mean(HRR20mm.HRR, 5, center=True), linewidth=2);

xlabel('Time / s')
ylabel('HRR / kw')
```

```python
Y=np.fft.fft(HRR20mm.HRR)
plot(Y)
ylim([0, 200])
```

```python
plot(HRR10mm.Time,HRR10mm.HRR, 'ko', markersize=1)

# average from 2 before until 2 after (window size = 5)
plot(HRR10mm.Time, pd.rolling_mean(HRR10mm.HRR, 5, center=True), linewidth=2);

xlabel('Time / s')
ylabel('HRR / kw')
```

```python
Y=np.fft.fft(HRR10mm.HRR)
plot(Y)
ylim([0, 200])
```

```python
plot(HRR5mm.Time,HRR5mm.HRR, 'ko', markersize=1)

# average from 2 before until 2 after (window size = 5)
plot(HRR5mm.Time, pd.rolling_mean(HRR5mm.HRR, 5, center=True), linewidth=1);

xlabel('Time / s')
ylabel('HRR / kw')
```

```python
Y=np.fft.fft(HRR5mm.HRR)
plot(Y)
ylim([0, 200])
```

```python
plot(HRR20mm.Time, pd.rolling_mean(HRR20mm.HRR, 5, center=True), 'b', linewidth=1, label="20mm");
plot(HRR10mm.Time, pd.rolling_mean(HRR10mm.HRR, 5, center=True), 'r', linewidth=1, label="10mm");
plot(HRR5mm.Time, pd.rolling_mean(HRR5mm.HRR, 5, center=True), 'g', linewidth=1, label="5mm");

legend(loc='lower right')
xlabel('Time / s')
ylabel('HRR / kw')
```

    From the Figures above, it is obviously that the HRR with different mesh sizes have slmost the same average value, but vibrate with different swings. The swings decrease with the mesh sizes. From the FFT analysis, the HRR didn't present any special vibration frequency, which means such vibration is just the noize from simulation.

## 2. Oxygen

```python
Dev20mm = pd.read_csv('/home/qize/FDS_simulation/Copy_sever/mesh_sensitivity_study/20mm/2D_CVC_V40_15SLPM_devc.csv', header=1)
Dev10mm = pd.read_csv('/home/qize/FDS_simulation/Copy_sever/mesh_sensitivity_study/10mm/2D_CVC_V40_15SLPM_devc.csv', header=1)
Dev5mm = pd.read_csv('/home/qize/FDS_simulation/Copy_sever/mesh_sensitivity_study/5mm/2D_CVC_V40_15SLPM_devc.csv', header=1)
```

```python
plot(Dev10mm.Time,Dev10mm.GAS_Oxygen, 'ok', markersize=1)
# average from 2 before until 2 after (window size = 5)
plot(Dev10mm.Time, pd.rolling_mean(Dev10mm.GAS_Oxygen, 5, center=True), linewidth=2);

xlabel('Time / s')
ylabel('Y_Oxygen / %')
```

```python
Y=np.fft.fft(Dev10mm.GAS_Oxygen)
plot(Y)
ylim([0, 2])

```

```python
plot(Dev20mm.Time, pd.rolling_mean(Dev20mm.GAS_Oxygen, 5, center=True), 'b', linewidth=1);
plot(Dev10mm.Time, pd.rolling_mean(Dev10mm.GAS_Oxygen, 5, center=True), 'r', linewidth=1);
plot(Dev5mm.Time, pd.rolling_mean(Dev5mm.GAS_Oxygen, 5, center=True), 'g', linewidth=1);

xlabel('Time / s')
ylabel('Y_Oxygen / %')
```

    From the figure above, it seems that the virbation of mesh size 10mm has greatly reduced when campared with mesh size 20mm, and is also campariable wih mesh size of 5mm. So it is reasonable to choose 10mm as the mesh size for Mixture Fraction Model simulation.
