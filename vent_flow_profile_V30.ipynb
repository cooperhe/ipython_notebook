```python
%pylab inline
import pandas as pd
D30_2700 = pd.read_csv('/home/qize/FDS_simulation/Copy_sever/V30/2700/2D_CVC_devc.csv', header=1)
D30_2250 = pd.read_csv('/home/qize/FDS_simulation/Copy_sever/V30/2250/2D_CVC_devc.csv', header=1)
D30_1800 = pd.read_csv('/home/qize/FDS_simulation/Copy_sever/V30/1800/2D_CVC_devc.csv', header=1)
D30_1350 = pd.read_csv('/home/qize/FDS_simulation/Copy_sever/V30/1350/2D_CVC_devc.csv', header=1)
D30_900 = pd.read_csv('/home/qize/FDS_simulation/Copy_sever/V30/900/2D_CVC_devc.csv', header=1)
D30_450 = pd.read_csv('/home/qize/FDS_simulation/Copy_sever/V30/450/2D_CVC_devc.csv', header=1)

QQ=array([21.2,17.7,14.1,10.6,7.1,3.1])
V30out=array([mean(D30_2700.FLOW_Out_v),mean(D30_2250.FLOW_Out_v),mean(D30_1800.FLOW_Out_v),mean(D30_1350.FLOW_Out_v)\
              ,mean(D30_900.FLOW_Out_v),mean(D30_450.FLOW_Out_v)])
V30in=array([mean(D30_2700.FLOW_In_v),mean(D30_2250.FLOW_In_v),mean(D30_1800.FLOW_In_v),mean(D30_1350.FLOW_In_v)\
              ,mean(D30_900.FLOW_In_v),mean(D30_450.FLOW_In_v)])

figure(figsize=(6,4))
plot(QQ,V30out,'ro', label='out flow rate')
plot(QQ,V30in,'bs',label='in flow rate')

xlabel('$\dot Q_f \,/\, kw$',size=20)
ylabel('$ \dot V \,/\, m^3/s$',size=20)
legend(loc='upper left', numpoints=1)
xlim(0,25)
grid()
```
