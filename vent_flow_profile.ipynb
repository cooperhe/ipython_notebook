```python
QQ=array([21.2,17.7,14.1,10.6,7.1,3.1])
VD=array([mean(DEVC_2700.FLOW_Out_v)/mean(DEVC_2700.FLOW_In_v),mean(DEVC_2250.FLOW_Out_v)/mean(DEVC_2250.FLOW_In_v)\
          ,mean(DEVC_1800.FLOW_Out_v)/mean(DEVC_1800.FLOW_In_v),mean(DEVC_1350.FLOW_Out_v)/mean(DEVC_1350.FLOW_In_v)\
          ,mean(DEVC_900.FLOW_Out_v)/mean(DEVC_900.FLOW_In_v),mean(DEVC_450.FLOW_Out_v)/mean(DEVC_450.FLOW_In_v)])

O2out=array([mean(DEVC_2700.FLOW_Out_Oxygen),mean(DEVC_2250.FLOW_Out_Oxygen),mean(DEVC_1800.FLOW_Out_Oxygen)\
             ,mean(DEVC_1350.FLOW_Out_Oxygen),mean(DEVC_900.FLOW_Out_Oxygen),mean(DEVC_450.FLOW_Out_Oxygen)])
O2in=array([mean(DEVC_2700.FLOW_In_Oxygen),mean(DEVC_2250.FLOW_In_Oxygen),mean(DEVC_1800.FLOW_In_Oxygen)\
             ,mean(DEVC_1350.FLOW_In_Oxygen),mean(DEVC_900.FLOW_In_Oxygen),mean(DEVC_450.FLOW_In_Oxygen)])

Tg=VD*298-273

from scipy.optimize import curve_fit
def func(x, a, b):
    return a*x+b

p1, cov1 = curve_fit(func, QQ, Tg)
yfit = func(QQ, p1[0], p1[1])

figure(figsize=(6,4))

plot(QQ,Tg,'ro')
plot(QQ,yfit,'b-')
xlabel('$\dot Q_f / kw$',size=20)
ylabel('$ T_g / ^oC$',size=20)
xlim(0,25)
ylim(0,600)
grid()

p1
```

##Fitting using the Gaussian profile:

$$\frac{v}{v_m}=exp[-(\frac{x}{b})^2]$$

```python
X = np.linspace(-0.4, 0., 41)
fig = plt.figure(figsize=(8,6))

from scipy.optimize import curve_fit
def func(x, vm, b,vin):
    return vm*exp(-((x)/b)**2)+vin
p1, cov1 = curve_fit(func, X, numpy.mean(mat_0['TT1'], axis=0))
vfit = func(X, p1[0],p1[1],p1[2])

plot(X,numpy.mean(mat_0['TT1'], axis=0),'bo',markevery=2, label='FDS-21.2kw')
plot(X,vfit,'r-', label='Gaussion fit')

xlim(-0.4,0)
#ERR_i=numpy.std(mat_i['TT1'], axis=0)
xlabel('$x / m$',size=20)
ylabel('$  u / m/s$',size=20)
legend(loc='upper left', numpoints=1)
fig.text(0.15, 0.6, r'$u=1.591exp[-(x/0.195)^2]-0.383$', size=18)
plot([-0.4,0],[0,0],'k', linewidth=2)
grid()
p1
```
