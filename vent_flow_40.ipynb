```python
%pylab inline

T0=25.
P=range(200)
Vi=range(200)
Vo=range(200)
i=0
for Tg in np.linspace(25,1000,200):
    e=2.*(T0-Tg)/(T0+Tg)
    D=0.319
    Pcri=0.2427*(1.+0.5*e)*exp(1.1072*e)*(4*9.8*D*1.21*(Tg-T0)/(Tg+273))
    Vcri=1.0/4.0*np.pi*D**2*(2*9.8*D*(Tg-T0)/(Tg+273))*0.1754*exp(0.5536*e)
    Vmax=0.055*sqrt(D**5.*9.8*(-e))
    m3=0.7070
    M=9.1


    for P0_to_Pcri in np.linspace(0,1.0,200):
        Vnet=Vcri*(M-(1.+(M**2-1.)*(1.-P0_to_Pcri))**0.5)/(M-1.)
        Vin=Vmax*((1.+m3/2.)*(1.-P0_to_Pcri)**2-(2.+m3/2.)*(1.-P0_to_Pcri))**2
        if 0.<(Vnet+Vin)/Vin-(Tg+273)/(T0+273)<1.:
            P[i]=P0_to_Pcri
            Vi[i]=Vin
            Vo[i]=Vin+Vnet           
            break
    i=i+1
    
figure(figsize=(12,4))
subplot(1,2,1)
plot(np.linspace(26,1000,200),P,'b-', label='P0_to_Pcri')
xlabel('$T_g / ^oC$')
ylabel('$ P_0 / P_{cri}$')
title('P0_to_Pcri predicted by Coopy')
legend(loc='lower center')
grid()
subplot(1,2,2)
plot(np.linspace(26,1000,200),Vi,'b-', label='Cooper-inflow rate')
plot(np.linspace(26,1000,200),Vo,'r-',label='Cooper-outflow rate')
xlabel('$T_g / ^oC$')
ylabel('$ flow rate / m^3/s$')
title('flow predicted by Coopy')
legend(loc='lower center')
grid()
```
