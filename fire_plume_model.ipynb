#Plume Model

***
##Heskestad Model

**Asumption**
1. Only consider $Q_c$
2. Weak plume theory, $\rho_\infty \approx \rho$
3. Same top hat profile whith velocity, temperature and species
4. $v=\alpha u$

<center>**Coutinuity:**   $ \frac{d}{dz} (\rho \pi b^2 u) = \rho 2 \pi b \alpha u$

<center>**Momentum:**   $\frac{d}{dz} (b^2 u^2)= b^2 g \frac{\rho_\infty - \rho}{\rho_\infty}$

<center>**Buoyancy:**   $\frac{d}{dz} (b^2ug \frac{\rho_\infty - \rho}{\rho_\infty})=0$

<center>**Energy:**   $\dot{Q} _c =\rho u \pi b^2 c_p (T-T_\infty) $

<center> $\Downarrow $
 
<center> **Solution:**
 
<center> b: $b= \frac{6\alpha}{5} z$
 
<center> u： $u=\frac{5}{6}\,(\frac{9}{10\pi \alpha^2})^\frac{1}{3} \,g^\frac{1}{3} \,(c_p \rho_\infty T_\infty)^{-\frac{1}{3}} \,\dot{Q} _c ^\frac{1}{3} \,z^{-\frac{1}{3}}$ 
 
<center> T: $\frac{ \triangle T}{T_\infty}=\frac{5}{6} \,(\frac{9\pi ^2 \alpha ^4}{10})^{-\frac{1}{3}} \,g^ {-\frac{1}{3}} \,(c_p T \infty \rho_\infty)^{-\frac{2}{3}} \,\dot{Q} _c ^\frac{2}{3} \,z^{-\frac{5}{3}}$ 
 
<center>  $\Downarrow $<p>

From Experiment correlation, the centerline values of mean excess temperture and mean velocity have been found
  
  
<center>  $b _{\triangle T}= 0.12 (\frac{T_0}{T_\infty})^{1/2}\ (z-z_0)$

<center>  $\triangle T_0=9.1 \,(\frac{T_\infty}{gc_p^2\rho_\infty^2})^\frac{1}{3} \,\dot{Q} _c ^\frac{2}{3} \,(z-z_0)^{-\frac{5}{3}}$ 
  
<center>  $u_0=3.4 \,(\frac{g}{c_p \rho_\infty T_\infty})^\frac{1}{3} \,\dot{Q} _c ^\frac{1}{3} \,(z-z_0)^{-\frac{1}{3}}$ 
  
<center>  $z_0=-1.02D+0.083\dot{Q} _c ^\frac{2}{5}$</center>
  
**Note:** $b _{\triangle T}$ is the plume radius to the point where the temperature rise has declined to $0.5\triangle T_0$
  
**ALL this above is for point source**

***
**In our condition the 2D model, the fire plume should be soon restricted by the compartment wall at the weight direction since the 3D compartment is flat. By Assumpt the uniform distributing of temperture and velocity in weight, we get:**
<p>
  
<center>**Coutinuity:**   $ \frac{d}{dz} (\rho b wu) = \rho w\alpha u$

<center>**Momentum:**   $\frac{d}{dz} (bw u^2)= bw g \frac{\rho_\infty - \rho}{\rho_\infty}-f$

<center>**Buoyancy:**   $\frac{d}{dz} (bw ug \frac{\rho_\infty - \rho}{\rho_\infty})=0$

<center>**Energy:**   $\dot{Q} _c =\rho u  bw c_p (T-T_\infty) $

<center> $\Downarrow $

This Equations should have the same kind solution of the infinited line fire models present by MaCaffrey.

[MaCaffrey](http://fire.nist.gov/bfrlpubs/fire79/PDF/f79001.pdf)

<center>  $b = \frac{2}{\sqrt{\pi}}\alpha (z-z_0)$

<center>  $\triangle T=\frac{\sqrt{2}}{2}\alpha^{-\frac{2}{3}} \,(\frac{T_\infty}{gc_p^2\rho_{\infty}^2 W^2})^\frac{1}{3} \,\dot{Q} _c ^\frac{2}{3} \,(z-z_0)^{-1}$ 
  
<center>  $u=\alpha^{-\frac{1}{3}} \,(\frac{g}{c_p \rho_\infty T_\infty W})^\frac{1}{3} \,\dot{Q} _c ^\frac{1}{3}$
